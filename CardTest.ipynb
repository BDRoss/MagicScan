```python
import sys
import numpy as np
import cv2
from os import listdir
from os import makedirs
from os import path
import matplotlib.pyplot as plt

# Define title section
#y1=60
#y2=110
#x1=60
#x2=580

y1=20
y2=36
x1=20
x2=230



if not path.isdir("titles"):
    makedirs("titles")
for f in listdir("_Template/test"):
    print(f)
    filename = "_Template/test/" + f
    img = cv2.imread(filename)
    # Grayscale
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    # Only title
    title = gray[y1:y2, x1:x2]
    # Threshold to remove background
    flag, thresh = cv2.threshold(title, 150, 255, cv2.THRESH_BINARY)
    filename2="titles/" + f
    plt.figure(figsize=[20,10])
    plt.subplot(121); plt.axis('off'); plt.imshow(thresh, cmap="gray"); plt.title("Original Card")
    cv2.imwrite(filename2, thresh)
```
