```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cardFileName = "./test_assets/test2.jpg"
print("Reading card image: ", cardFileName)

# To be quite frank, I'm not sure why we read in the image, then convert it to RGB when it already should be?
card = cv2.imread(cardFileName, cv2.IMREAD_COLOR)
card = cv2.cvtColor(card, cv2.COLOR_BGR2RGB)

plt.figure(figsize=[20,10]);
plt.subplot(121); plt.axis('off'); plt.imshow(card); plt.title("Original Card")
```
